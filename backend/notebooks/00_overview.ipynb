## ğŸ—‚ï¸ é¡¹ç›®ç»“æ„æ¦‚è§ˆ

````python
# æŸ¥çœ‹é¡¹ç›®ç»“æ„
import os

def print_tree(directory, prefix="", max_depth=2, current_depth=0):
    """æ‰“å°ç›®å½•æ ‘"""
    if current_depth >= max_depth:
        return
    
    try:
        items = sorted(os.listdir(directory))
        # è¿‡æ»¤æ‰éšè—æ–‡ä»¶å’Œç‰¹å®šç›®å½•
        items = [item for item in items if not item.startswith('.') 
                 and item not in ['__pycache__', 'node_modules', 'venv', '.venv']]
        
        for i, item in enumerate(items):
            path = os.path.join(directory, item)
            is_last = i == len(items) - 1
            
            print(f"{prefix}{'â””â”€â”€ ' if is_last else 'â”œâ”€â”€ '}{item}")
            
            if os.path.isdir(path):
                extension = "    " if is_last else "â”‚   "
                print_tree(path, prefix + extension, max_depth, current_depth + 1)
    except PermissionError:
        pass

print("\nBackend é¡¹ç›®ç»“æ„:")
print("="*60)
print(backend_dir.name + "/")
print_tree(backend_dir, max_depth=2)
print("="*60)
````
